```python
df1 = pd.DataFrame({
    "key": ["b", "b", "a", "c", "a", "a", "b"],
    "data1": range(7)
})
df2 = pd.DataFrame({
    "key": ["a", "b", "d"],
    "data2": range(3)
})
```

```python
df1
```

```python
df2
```

#### 如果不指定列名，merge会自动将重叠列名作为连接的键

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="key")
```

#### 如果列名不同，可以分别指定列名

```python
df3 = pd.DataFrame({
    "lkey": ["b", "b", "a", "c", "a", "a", "b"],
    "data1": range(7)
})
df4 = pd.DataFrame({
    "rkey": ["a", "b", "d"],
    "data2": range(3)
})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on="lkey", right_on="rkey")
```

#### merge默认使用inner join，结果为两张表的交集；如果想取并集，指定how为outer即可

```python
pd.merge(df1, df2, how="outer")
```

```python
df1 = pd.DataFrame({
    "key": ["b", "b", "a", "c", "a", "b"],
    "data1": range(6)
})
df2 = pd.DataFrame({
    "key": ["a", "b", "a", "b", "d"],
    "data2": range(5)
})
```

```python
df1
```

```python
df2
```

#### 多对多连接是行的笛卡尔积

```python
pd.merge(df1, df2, on="key", how="left")
```

```python
left = pd.DataFrame({
    "key1": ["foo", "foo", "bar"],
    "key2": ["one", "two", "one"],
    "lval": [1, 2, 3]
})
right = pd.DataFrame({
    "key1": ["foo", "foo", "bar", "bar"],
    "key2": ["one", "one", "one", "two"],
    "rval": [4, 5, 6, 7]
})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=["key1", "key2"], how="outer")
```

```python
pd.merge(left, right, on="key1")
```

```python
pd.merge(left, right, on="key1", suffixes=("_left", "_right"))
```

```python
pd.merge(left1, right1, left_on="key", right_index=True)
```
