### Read & Write Database

```python
import sqlite3

conn = sqlite3.connect("test.db")
cursor = conn.cursor()

cursor.execute("CREATE TABLE IF NOT EXISTS user (id VARCHAR(20) PRIMARY KEY, name VARCHAR(20))")
cursor.execute("INSERT INTO user (id, name) values ('1', 'Michael')")
cursor.execute("INSERT INTO user (id, name) values ('2', 'Jordan')")

conn.commit()

cursor.execute("SELECT * FROM user WHERE id = ? and name = ?", ("2", "Jordan"))
res = cursor.fetchall()
print(res)

cursor.close()
conn.close()
```

```python
conn = sqlite3.connect("test.db")

cursor = conn.cursor()
cursor.execute("INSERT INTO user (id, name) values ('3', 'Jay'), ('4', 'Sherry'), ('5', 'Alex'), ('6', 'Kitty')")

conn.commit()
```

```python
cursor.execute("SELECT * FROM user")
res = cursor.fetchall()
for r in res:
    print(r)

cursor.close()
conn.close()
```

```python
def get_conn(db_name):
    return sqlite3.connect(db_name)

def get_cursor(conn):
    return conn.cursor()
    
def close_conn(conn, cursor):
    if cursor is not None:
        cursor.close()
    if conn is not None:
        conn.close()
```

```python
conn = get_conn("test.db")
res = pd.read_sql("SELECT * FROM user", con=conn)
close_conn(conn, None)
display(res)
```

```python
conn = get_conn("test.db")
df5.to_sql("abc", con=conn)
res = pd.read_sql("SELECT * FROM abc", con=conn)
close_conn(conn, None)
display(res)
```

```python
jobs = pd.read_csv(r"./09-job_info.csv", sep=";", header=None, names=["user", "application_id", "application_type", "start_time", "end_time", "duration", "cpu_time", "memory_time"])
```

```python
jobs
```

```python
jobs.info()
```

```python
import time

def timestamp_to_localtime(timestamp):
    return time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(timestamp))
```

```python
jobs["start_time"] = (jobs["start_time"] / 1000).map(timestamp_to_localtime)
jobs["end_time"] = (jobs["end_time"] / 1000).apply(timestamp_to_localtime)
```

```python
jobs["duration"] = jobs["duration"] / 3600000
```
