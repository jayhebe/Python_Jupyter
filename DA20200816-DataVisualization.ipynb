### 绘制子图

```python
# 创建画布对象
figure = plt.figure()

# 添加子图
figure.add_subplot(1, 2, 1, facecolor="g") # 整个画布分为几行，几列，当前绘制区域
plt.plot(10, marker="o")

# 添加子图
figure.add_subplot(1, 2, 2, facecolor="gray") # 整个画布分为几行，几列，当前绘制区域
plt.plot(20, marker="d")

plt.subplots_adjust(wspace=1)
```

```python
# 创建画布对象
figure = plt.figure()

# 添加子图
sub1 = figure.add_subplot(1, 2, 1, facecolor="g") # 整个画布分为几行，几列，当前绘制区域

# 添加子图
sub2 = figure.add_subplot(1, 2, 2, facecolor="gray") # 整个画布分为几行，几列，当前绘制区域


sub2.plot(20, marker="d")
sub1.plot(10, marker="o")

plt.subplots_adjust(wspace=1)
```

### subplot方法
* 会隐式创建画布

```python
plt.subplot(1, 2, 1)
plt.plot([3, 8, 6], "g--d")

plt.subplot(1, 2, 2)
plt.plot([3, 8, 4], "r-.o")
```

### subplots方法

```python
figure, ax = plt.subplots(2, 2, sharex=True, sharey=True)
ax[0][0].plot([3, 9])
ax[0][1].plot([3, 9])
ax[1][0].plot([3, 9])
ax[1][1].plot([3, 9])
```

```python
figure, ax = plt.subplots(2, 2, sharex=True, sharey=True)
ax[0][0].plot([3, 9])
ax[0][1].plot([3, 9])
ax[1][0].plot([3, 9])
ax[1][1].plot([3, 9])

figure.set_size_inches(10, 10)
```

### 标签与刻度

```python
plt.plot([1, 3, 5], [2, 4, 6])
print(plt.xlim())
print(plt.ylim())

# Set x scope
plt.xlim(0, 10)

# Get x scale
plt.xticks()

# Set x scale
plt.xticks([0, 5, 10])

# Set x label
plt.xticks([0, 5, 10], labels=["low", "medium", "high"])
```

```python
# axis
plt.axis()

# plt.axis((xmin, xmax, ymin, ymax))
plt.axis((0, 10, -5, 10))
plt.plot([1, 2, 3], [1, 10, 26])
```

```python
plt.axis("equal")
plt.plot([1, 2, 3], [1, 10, 26])
```
