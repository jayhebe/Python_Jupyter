```python
meal = pd.read_csv("./data/meal_order_info.csv", encoding="gbk")
```

```python
meal.info()
```

```python
meal.head().T
```

```python
meal["lock_time"] = pd.to_datetime(meal["lock_time"])
```

```python
date_index = pd.DatetimeIndex(meal["lock_time"])
date_index
```

```python
meal["lock_time"] = date_index
meal["lock_time"]
```

```python
# freq = S - second, T - Minute, H - Hour, D - Day, M - Month, Y - Year

period_index = pd.PeriodIndex(meal["lock_time"], freq="T")
period_index
```

```python
lock_year = [y.year for y in meal["lock_time"]]
lock_month = [y.month for y in meal["lock_time"]]
lock_day = [y.day for y in meal["lock_time"]]
lock_weekday = [y.weekday() for y in meal["lock_time"]]
display(lock_year[:5])
display(lock_month[:5])
display(lock_day[:5])
display(lock_weekday[-6:-1])
```

```python
time_delta = meal["lock_time"] + pd.Timedelta(weeks=-1)
time_delta
```

```python
time_delta2 = meal["lock_time"] - pd.to_datetime("2020-08-08")
time_delta2
```

```python
meal["use_start_time"] = pd.to_datetime(meal["use_start_time"])
```

```python
timemin = meal["lock_time"].min()
timemax = meal["lock_time"].max()
display(timemin)
display(timemax)
display(timemax - timemin)
```

```python
check_time = meal["lock_time"] - meal["use_start_time"]
display(check_time.mean())
display(check_time.min())
display(check_time.max())
```
