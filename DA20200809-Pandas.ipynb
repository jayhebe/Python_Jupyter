```python
# freq = S - second, T - Minute, H - Hour, D - Day, M - Month, Y - Year

period_index = pd.PeriodIndex(meal["lock_time"], freq="T")
period_index
```

```python
lock_year = [y.year for y in meal["lock_time"]]
lock_month = [y.month for y in meal["lock_time"]]
lock_day = [y.day for y in meal["lock_time"]]
lock_weekday = [y.weekday() for y in meal["lock_time"]]
display(lock_year[:5])
display(lock_month[:5])
display(lock_day[:5])
display(lock_weekday[-6:-1])
```

```python
timemin = meal["lock_time"].min()
timemax = meal["lock_time"].max()
display(timemin)
display(timemax)
display(timemax - timemin)
```
