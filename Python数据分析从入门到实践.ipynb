```python
df = pd.read_excel(r"Code/07/example/02/01/JD2019.xlsx")
df = df.set_index("日期")
df_etl = pd.concat([df["2019-02"], df["2020-02"]])
# 以下两个操作效果相同
# df_etl = df_etl[df_etl["商品名称"] == "零基础学Python（全彩版）"]
df_etl = df_etl.query("商品名称 == '零基础学Python（全彩版）'")
df1 = df_etl[["北京", "上海", "广州", "成都", "武汉", "沈阳", "西安"]]
df2 = df1.T
x = np.arange(7)
y1 = df2["2019-02-01"]
y2 = df2["2020-02-01"]
df2["rate"] = ((y2 - y1) / y1) * 100
width = 0.25
plt.figure(figsize=(10, 6))
plt.rcParams["font.sans-serif"] = ["Microsoft YaHei"]
plt.title("全国各地区销量及同比增长情况")
plt.ylabel("销售数量（册）")
plt.xticks(x, ["北京", "上海", "广州", "成都", "武汉", "沈阳", "西安"])
plt.bar(x, y1, width=width, color="orange", label="2019年2月")
plt.bar(x + width, y2, width=width, color="deepskyblue", label="2020年2月")
for a, b in zip(x, df2["rate"]):
    plt.text(a, b, ("%.1f%%" % b), ha="center", va="bottom", fontsize=11)
plt.legend()
plt.show()
```

```python
df = pd.read_excel(r"Code/07/example/02/02/JD2019.xlsx")
df1 = df.query("商品名称 == '零基础学Python（全彩版）'").sort_values("日期")
df1 = df1[["北京", "上海", "广州", "成都", "武汉", "沈阳", "西安", "日期"]]
df1 = df1.set_index("日期")
df1["全国销量"] = df1.sum(axis=1)
df1 = df1["2019-01-01":"2019-12-01"]
df1["January"] = df1.iloc[0, 7]
df1["base"] = df1["全国销量"] / df1["January"]
x = np.arange(12)
y1 = df1["全国销量"]
y2 = df1["base"]
fig = plt.figure(figsize=(10, 6))
plt.rcParams["font.sans-serif"] = ["Microsoft YaHei"]
ax1 = fig.add_subplot(111)
plt.title("2019年全国销量定比分析")
plt.xticks(x, [str(s) + "月" for s in range(1, 13)])
ax1.bar(x, y1, color="blue", label="left", alpha=0.5)
ax1.set_ylabel("全国销量（册）")
ax2 = ax1.twinx()
ax2.plot(x, y2, color="r", linestyle="-", marker="D", linewidth=2)
for a, b in zip(x, y2):
    plt.text(a, b + 0.02, "%.3f" % b, ha="center", va="bottom", fontsize=9)
plt.show()
```
