```python
def feature_equals(df):
    df_equals = pd.DataFrame([], columns=df.columns, index=df.columns)
    for i in df.columns:
        for j in df.columns:
            df_equals.loc[i, j] = df.loc[:,i].equals(df.loc[:,j])
    return df_equals
```

```python
def out_range(s):
    qu = s.quantile(0.75)
    ql = s.quantile(0.25)
    iqr = qu - ql
    s.loc[s > qu + 1.5 * iqr] = qu
    s.loc[s < ql - 1.5 * iqr] = ql
    
    return s
```

```python
print(detail["counts"].max(), detail["counts"].min())
print(detail["amounts"].max(), detail["amounts"].min())
```

```python
def min_max_scale(data):
    data = (data - data.min()) / (data.max() - data.min())
    return data
```

```python
d1 = min_max_scale(detail["counts"])
d2 = min_max_scale(detail["amounts"])

display(detail[["counts", "amounts"]].head())
display(pd.concat([d1, d2], axis=1).head())
```

```python
def z_score(data):
    data = (data - data.mean()) / data.std()
    return data
```

```python
d3 = z_score(detail["counts"])
d4 = z_score(detail["amounts"])

display(detail[["counts", "amounts"]].head())
display(pd.concat([d3, d4], axis=1).head())
```

```python
def decimal_scaler(data):
    k = np.ceil(np.log10(data.abs().max()))
    data = data / 10 ** k
    return data
```

```python
d5 = decimal_scaler(detail["counts"])
d6 = decimal_scaler(detail["amounts"])

display(detail[["counts", "amounts"]].head())
display(pd.concat([d5, d6], axis=1).head())
```

### kmeans聚类
* Need scikit-learn module
* pip install scikit-learn -i https://mirrors.aliyun.com/pypi/simple/

```python
def kmeans_cut(data, k):
    from sklearn.cluster import KMeans
    km = KMeans(n_clusters=k)
    km.fit(data.reshape((len(data), 1)))
    
    c = pd.DataFrame(km.cluster_centers).sort_values(0)
    w = pd.Series.rolling(c, 2).iloc[1:]
    w = [0] + list(w[0]) + [data.max()]
    data = pd.cut(data, w)
    
    return data
```
